```python
#btvn_a*
import heapq

def tsp_a_star(matrix):
    n = len(matrix)
    # Trạng thái: (f_score, current_city, visited_mask, g_score, path)
    # visited_mask dùng bitmask để tiết kiệm bộ nhớ
    start_node = (0, 0, 1, 0, [0])
    open_set = [start_node]

    # Lưu g_score tốt nhất cho mỗi trạng thái (vị trí, tập đã thăm)
    visited_states = {}

    while open_set:
        f, u, mask, g, path = heapq.heappop(open_set)

        # Nếu đã thăm hết các thành phố
        if mask == (1 << n) - 1:
            # Cộng thêm đường về đỉnh 0
            return path + [0], g + matrix[u][0]

        state_key = (u, mask)
        if state_key in visited_states and visited_states[state_key] <= g:
            continue
        visited_states[state_key] = g

        for v in range(n):
            if not (mask & (1 << v)): # Nếu v chưa thăm
                new_g = g + matrix[u][v]
                new_mask = mask | (1 << v)

                # Heuristic đơn giản: cạnh nhỏ nhất còn lại (Min Edge)
                # Để tối ưu hơn, có thể dùng MST
                h = calculate_heuristic(matrix, v, new_mask, n)

                heapq.heappush(open_set, (new_g + h, v, new_mask, new_g, path + [v]))
    return None

def calculate_heuristic(matrix, u, mask, n):
    # Ước lượng: Khoảng cách từ u đến thành phố gần nhất chưa thăm
    # cộng với khoảng cách từ một thành phố chưa thăm về đích
    remaining = [i for i in range(n) if not (mask & (1 << i))]
    if not remaining:
        return matrix[u][0]

    # Heuristic đơn giản: Lấy min các cạnh còn lại
    res = min(matrix[u][i] for i in remaining)
    return res

# Dữ liệu mẫu (Ma trận kề)
dist_matrix = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

path, total_cost = tsp_a_star(dist_matrix)
print(f"Đường đi tối ưu: {path}")
print(f"Tổng chi phí: {total_cost}")
```
