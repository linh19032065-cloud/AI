```python
#bttl a*
import heapq

def a_star(graph, start, goal, h):
    """
    graph: Danh sách kề {đỉnh: [(đỉnh_kề, trọng_số), ...]}
    start: Đỉnh bắt đầu
    goal: Đỉnh kết thúc
    h: Từ điển chứa giá trị Heuristic {đỉnh: giá trị}
    """

    # Hàng đợi ưu tiên lưu (f_score, node)
    open_list = []
    heapq.heappush(open_list, (h[start], start))

    # Lưu vết đường đi
    came_from = {}

    # g_score[n]: chi phí thực tế từ start đến n
    g_score = {node: float('inf') for node in graph}
    g_score[start] = 0

    while open_list:
        # Lấy đỉnh có f_score = g + h thấp nhất
        current_f, current = heapq.heappop(open_list)

        # Nếu đã đến đích, truy vết đường đi và trả về
        if current == goal:
            return reconstruct_path(came_from, current), g_score[goal]

        for neighbor, weight in graph.get(current, []):
            # Tính g_score mới cho đỉnh kề
            tentative_g_score = g_score[current] + weight

            if tentative_g_score < g_score[neighbor]:
                # Đây là đường đi tốt nhất tìm thấy cho đến nay
                came_from[neighbor] = current
                g_score[neighbor] = tentative_g_score
                f_score = tentative_g_score + h.get(neighbor, 0)

                # Nếu chưa có trong hàng đợi, thêm vào
                if not any(neighbor == item[1] for item in open_list):
                    heapq.heappush(open_list, (f_score, neighbor))

    return None, float('inf')

def reconstruct_path(came_from, current):
    path = [current]
    while current in came_from:
        current = came_from[current]
        path.append(current)
    return path[::-1]

# --- VÍ DỤ SỬ DỤNG ---
if __name__ == "__main__":
    # 1. Khai báo đồ thị (Danh sách kề)
    adj_list = {
        'A': [('B', 1), ('C', 3)],
        'B': [('D', 3), ('E', 1)],
        'C': [('F', 2)],
        'D': [],
        'E': [('G', 2)],
        'F': [('G', 1)],
        'G': []
    }

    # 2. Khai báo hàm Heuristic (Ước lượng khoảng cách đến G)
    h_values = {
        'A': 6, 'B': 4, 'C': 3,
        'D': 5, 'E': 2, 'F': 1, 'G': 0
    }

    path, cost = a_star(adj_list, 'A', 'G', h_values)

    if path:
        print(f"Đường đi tối ưu: {' -> '.join(path)}")
        print(f"Tổng chi phí: {cost}")
    else:
        print("Không tìm thấy đường đi!")
```
